##1. Introduction
One popular way to consopeople look for ways to consolidate debt to reduce their monthly expenditures.  One popular way to consolidate debt is through P2P (Peer-to-Peer/Peer-2-Peer) loan platforms such as Prosper Marketplace.

P2P loan platforms, such as Prosper Marketplace, use a loanee's financial health attributes which get fed into a platforms's proprietary calculation to calculate a general financial health rating and if the loanee's loan request is worthy of being placed on the platform's open note market. which is used to calculate if a loanee is worthy of a loan at a particular APR (Annual Percentage Rate)

On the other hand, there are also people look into investing in loan notes via P2P lending platforms, in which investors generally look at financial health variables to get a general understanding if a note is worth their investment.

This research will be focused on what financial health attributes makes a successful note, were investors get their return on investment vs that of unsuccesful note, were there loan is charged off and investors lose a portion of their investment.   

Attributes such as a loanee's Debt-toIncome ratio, loanee's Fico and Scorex scores, and income range.  By pinpointing successful financial health attributes, readers who are looking to invest in notes on a platform such as Prosper may utilize findings when buying notes on a P2P lending platform.
